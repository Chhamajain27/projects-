# find the total amount invested by each shark

```python
l=[]
l1=[]
for i in df.columns[18:25]:
    x=df[df[i]==1]
    s=x['amount_per_shark'].sum()
    l.append(s)
    i=i.replace('_deal','')
    print(i,"invested",s,'amount')
    l1.append(i)
amount=sorted(l,reverse=True)
amount
```

```python
l=[]
l1=[]
for i in df.columns[18:25]:
    x=df[df[i]==1]
    s=x['equity_per_shark'].sum()
    l.append(s)
    i=i.replace('_deal','')
    print(i,"invested",s,'amount')
    l1.append(i)
amount=sorted(l,reverse=True)
amount
```

```python
df=pd.DataFrame({"shark name":l1,'Total equity':l})
df
```
